Remappage des IDs de classes dans les annotations YOLO pour chaque jeu (train, val, test) : lecture de chaque fichier .txt, remplacement de l‚Äôancien ID par le nouveau selon le dictionnaire remap, puis r√©√©criture du fichier avec les classes corrig√©es

```python
import os

# Mapping ancien ‚Üí nouveau
remap = {
    1: 0,
    2: 1,
    3: 2,
    4: 3,
    6: 4,
    7: 5,
    8: 6,
    9: 7
}

# Racine du dataset
base = "/content/drive/MyDrive/Last_Train/data"
splits = ["train", "val", "test"]

for split in splits:
    label_dir = os.path.join(base, split, "labels")
    for fname in os.listdir(label_dir):
        if fname.endswith(".txt"):
            path = os.path.join(label_dir, fname)
            new_lines = []
            with open(path, "r") as f:
                for line in f:
                    parts = line.strip().split()
                    if parts:
                        old_id = int(parts[0])
                        new_id = remap.get(old_id, None)
                        if new_id is not None:
                            new_line = " ".join([str(new_id)] + parts[1:])
                            new_lines.append(new_line)
            # √âcrase le fichier avec les classes corrig√©es
            with open(path, "w") as f:
                f.write("\n".join(new_lines))
```
